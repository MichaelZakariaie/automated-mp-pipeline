```python
def transform_df_to_single_row(df):
    """
    Transform dataframe into a single row with columns named trial[X]_[feature]
    where X is the trial number and feature is one of the specified columns.
    """
    # Define the columns we want to transform (per trial)
    trial_columns = [
        'dot_latency_pog', 'cue_latency_pog', 'num_unique_sacc', 
        'trial_saccade_data_quality_pog', 'cue_latency_pog_good', 
        'dot_latency_pog_good', 'percent_bottom_freeface_pog', 
        'percent_top_freeface_pog', 'fixation_quality'
    ]
    
    # Define columns that don't have trial numbers (single values)
    single_columns = ['percent_loss_late', 'session_saccade_data_quality_pog']
    
    # Create mapping for trial_saccade_data_quality_pog
    quality_mapping = {val: idx for idx, val in enumerate(df['trial_saccade_data_quality_pog'].unique())}
    
    # Create new dataframe structure
    new_data = {}
    
    # Add trial-based columns
    for idx, row in df.iterrows():
        trial_num = row['trial']
        
        for col in trial_columns:
            if col in df.columns:
                if col == 'trial_saccade_data_quality_pog':
                    # Label encode this column (convert to numbers)
                    col_name = f'trial{trial_num}_{col}'
                    new_data[col_name] = quality_mapping[row[col]]
                else:
                    col_name = f'trial{trial_num}_{col}'
                    new_data[col_name] = row[col]
    
    # Add single columns (take first occurrence or aggregate as needed)
    for col in single_columns:
        if col in df.columns:
            new_data[col] = df[col].iloc[0]  # Take first value
    
    # Create single-row dataframe
    result_df = pd.DataFrame([new_data])
    
    return result_df

# Apply the transformation
transformed_df = transform_df_to_single_row(df)
print(f"Transformed dataframe shape: {transformed_df.shape}")
print(f"Quality mapping: {dict(enumerate(df['trial_saccade_data_quality_pog'].unique()))}")
```

```python
# FINAL CORRECTED VERSION - 140 trials, excluding num_unique_sacc and dot_latency_pog_good
def transform_df_to_single_row_final(df):
    """
    Transform dataframe into a single row with columns named trial[X]_[feature]
    where X is the trial number and feature is one of the specified columns.
    Creates columns for ALL 140 trials (0-139), with NaN for missing trials.
    """
    # Define the columns we want to transform (per trial) - EXCLUDING num_unique_sacc and dot_latency_pog_good
    trial_columns = [
        'dot_latency_pog', 
        'cue_latency_pog', 
        'trial_saccade_data_quality_pog', 
        'cue_latency_pog_good', 
        'percent_bottom_freeface_pog', 
        'percent_top_freeface_pog',
        'fixation_quality'
    ]
    # Note: We are EXCLUDING: num_unique_sacc, dot_latency_pog_good
    
    # Define columns that don't have trial numbers (single values)
    single_columns = ['percent_loss_late', 'session_saccade_data_quality_pog']
    
    # Create mapping for trial_saccade_data_quality_pog
    quality_mapping = {val: idx for idx, val in enumerate(df['trial_saccade_data_quality_pog'].unique())}
    
    # Create new dataframe structure
    new_data = {}
    
    # We need to create columns for ALL 140 trials (0-139), not just the ones in the data
    for trial_num in range(140):  # 0 to 139
        for col in trial_columns:
            if col in df.columns:
                # Find if this trial exists in the data
                trial_row = df[df['trial'] == trial_num]
                
                if col == 'trial_saccade_data_quality_pog':
                    # Label encode this column (convert to numbers)
                    col_name = f'trial{trial_num}_{col}'
                    if not trial_row.empty:
                        new_data[col_name] = quality_mapping[trial_row.iloc[0][col]]
                    else:
                        new_data[col_name] = np.nan  # Missing trial
                else:
                    col_name = f'trial{trial_num}_{col}'
                    if not trial_row.empty:
                        new_data[col_name] = trial_row.iloc[0][col]
                    else:
                        new_data[col_name] = np.nan  # Missing trial
    
    # Add single columns (take first occurrence or aggregate as needed)
    for col in single_columns:
        if col in df.columns:
            new_data[col] = df[col].iloc[0]  # Take first value
    
    # Create single-row dataframe
    result_df = pd.DataFrame([new_data])
    
    return result_df

# Apply the transformation
transformed_df = transform_df_to_single_row_final(df)
print(f"Final transformed dataframe shape: {transformed_df.shape}")
print(f"Expected: 7 trial columns  140 trials + 2 single columns = {7 * 140 + 2}")

# Verify missing trials have NaN
print(f"\nValues for missing trial 13:")
missing_cols = [col for col in transformed_df.columns if 'trial13_' in col]
for col in missing_cols[:3]:
    print(f"  {col}: {transformed_df[col].iloc[0]}")

print(f"\nQuality mapping: {dict(enumerate(df['trial_saccade_data_quality_pog'].unique()))}")
print("0 = good, 1 = bad")
```

```python
# Process all parquet files in the analytical_sessions folder
import os
import glob

# Get all parquet files
parquet_files = glob.glob('analytical_sessions/*.parquet')
print(f"Found {len(parquet_files)} parquet files")

# Process each file
all_transformed_dfs = []

for parquet_file in parquet_files:
    print(f"\nProcessing: {os.path.basename(parquet_file)}")
    
    # Read the parquet file
    df_temp = pd.read_parquet(parquet_file)
    print(f"  Original shape: {df_temp.shape}")
    
    # Transform to single row using our function
    transformed_row = transform_df_to_single_row_corrected(df_temp)
    
    # Add source file identifier
    transformed_row['source_file'] = os.path.basename(parquet_file)
    transformed_row['session_id'] = os.path.basename(parquet_file).split('_')[0]
    
    all_transformed_dfs.append(transformed_row)

# Combine all transformed dataframes
combined_df = pd.concat(all_transformed_dfs, ignore_index=True)

# Reorder columns to put identifiers first
cols = combined_df.columns.tolist()
id_cols = ['session_id', 'source_file']
data_cols = [col for col in cols if col not in id_cols]
combined_df = combined_df[id_cols + data_cols]

print(f"\nFinal combined dataframe shape: {combined_df.shape}")
print(f"Columns: {len(id_cols)} identifier columns + {len(data_cols)} data columns = {len(combined_df.columns)} total")

# Save the results
combined_df.to_parquet('combined_analytical_sessions.parquet', index=False)
combined_df.to_csv('combined_analytical_sessions.csv', index=False)
print("\nSaved to 'combined_analytical_sessions.parquet' and '.csv'")
```
